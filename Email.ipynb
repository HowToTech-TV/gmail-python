```python
#Sending gmail
#Smtplib is a builtin module that help us send emails through the smtp protocol.

import smtplib
#insert mime objects
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

#login with app password (Refer to the first part of video)
username = input("Please enter user name:")
password = input("Please enter app password:")
sendto = input("Please enter recipient")
#create message using MIMEMultipart
message = MIMEMultipart("alternative")
message["From"] = username
message["To"] = sendto
message["Subject"] = input("Please enter subject: ")

#Let user input the main message
text = input('Please enter message: ')

mimetext = MIMEText(text, "text") 
#insert as plain text(html also supported)
message.attach(mimetext)
#gmail server port
server = smtplib.SMTP_SSL("smtp.gmail.com", 465)#port 465 for secure socket layer connection
try:
  server.login(username, password)
  #sending the email syntax: sendmail(from, to, message)
  server.sendmail(username, sendto, message.as_string())
  server.close()
except:
  print("Failed")
```
